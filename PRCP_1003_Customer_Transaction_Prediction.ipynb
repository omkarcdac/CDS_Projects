```python
X=data.iloc[:,2:]
y=data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier=RandomForestClassifier(random_state=42,n_estimators=100,n_jobs=-1,max_depth=5,oob_score=True)
rf_classifier.fit(X_train,y_train)
```

```python
rf_classifier.oob_score_
```

```python
rf=RandomForestClassifier(random_state=42,n_jobs=1)
params={
    'max_depth':[2,3,5,7,9,10],
     'min_samples_leaf':[100,200,300,400,500,600],
    'n_estimators':[100,200,300,400,500,600]
}
from sklearn.model_selection import GridSearchCV
gscv=GridSearchCV(rf,param_grid=params,cv=5,n_jobs=-1,verbose=1,scoring='accuracy')
gscv.fit(X_train,y_train)
```

```python
gscv.best_score_
```

```python
rf_best_model=gscv.best_estimator_
```

```python
rf_best_model
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy=accuracy_score(y_test,y_pred)
```
